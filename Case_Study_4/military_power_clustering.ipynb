## Deal with missing values 
* Hint-1, use wikipedia to fill in missing values comparing with similar countries
* Hint-2, Georeferences of countries can help you

```python
value_df[value_df["Geography_Total Coastline Coverage value"].isnull()]
```

```python
# as we can see when (Geography_Total Coastline Coverage value) column is NaN, other columns related to navy are mostly NaN
# I found that this countries have not coastline so this values will be assign 0
```

```python
value_df.loc[value_df["Geography_Total Coastline Coverage value"].isnull(), 
         ["Navy Strength_Navy Fleet Strengths value",
         "Navy Strength_Aircraft Carrier Fleet Strength value",
         "Navy Strength_Submarine Fleet Strength value",
         "Navy Strength_Destroyer Fleet Strength value",
         "Navy Strength_Frigate Fleet Strength value"]] = 0
```

```python
value_df["Geography_Total Coastline Coverage value"].fillna(value=0, inplace=True)
```

```python
value_df[(value_df["Navy Strength_Submarine Fleet Strength value"].isnull())|
         (value_df["Navy Strength_Frigate Fleet Strength value"].isnull())|
         (value_df["Navy Strength_Destroyer Fleet Strength value"].isnull())|
        (value_df["Navy Strength_Aircraft Carrier Fleet Strength value"].isnull())]

```

```python
value_df.loc[value_df.country == "Uzbekistan", 
            ["Navy Strength_Navy Fleet Strengths value",
             "Navy Strength_Aircraft Carrier Fleet Strength value",
             "Navy Strength_Submarine Fleet Strength value",
             "Navy Strength_Destroyer Fleet Strength value",
             "Navy Strength_Frigate Fleet Strength value"]] = 0
```

```python
value_df[(value_df["Navy Strength_Submarine Fleet Strength value"].isnull())|
         (value_df["Navy Strength_Frigate Fleet Strength value"].isnull())|
         (value_df["Navy Strength_Destroyer Fleet Strength value"].isnull())|
        (value_df["Navy Strength_Aircraft Carrier Fleet Strength value"].isnull())]

```

```python
value_df.loc[(value_df["Navy Strength_Navy Fleet Strengths value"] > 200) & (value_df["Navy Strength_Navy Fleet Strengths value"] < 250)]
```

```python
value_df.loc[110, "Navy Strength_Aircraft Carrier Fleet Strength value"] = (round(value_df.loc[(value_df["Navy Strength_Navy Fleet Strengths value"] > 200) &
              (value_df["Navy Strength_Navy Fleet Strengths value"] < 250)].mean())
  ["Navy Strength_Aircraft Carrier Fleet Strength value"])
```

```python
value_df.loc[110, "Navy Strength_Submarine Fleet Strength value"] = (round(value_df.loc[(value_df["Navy Strength_Navy Fleet Strengths value"] > 200) &
              (value_df["Navy Strength_Navy Fleet Strengths value"] < 250)].mean())
  ["Navy Strength_Submarine Fleet Strength value"])
```

```python
value_df.loc[110, "Navy Strength_Destroyer Fleet Strength value"] = (round(value_df.loc[(value_df["Navy Strength_Navy Fleet Strengths value"] > 200) &
              (value_df["Navy Strength_Navy Fleet Strengths value"] < 250)].mean())
  ["Navy Strength_Destroyer Fleet Strength value"])
```

```python
value_df.loc[110, "Navy Strength_Frigate Fleet Strength value"] = (round(value_df.loc[(value_df["Navy Strength_Navy Fleet Strengths value"] > 200) &
              (value_df["Navy Strength_Navy Fleet Strengths value"] < 250)].mean())
  ["Navy Strength_Frigate Fleet Strength value"])
```
